# DF.Missing Data

```python
import pandas as pd
import numpy as np
```

```python
a=pd.DataFrame({'A':[11,34,np.nan,12,np.nan,35],
                'B':[22,32,25,15,9,29],
                'C':[23,np.nan,24,16,19,22],
                'D':[np.nan,np.nan,25,33,18,np.nan],
                'E':[12,13,24,26,30,32]})
```

```python
a
```

```python
a.dropna()
```

```python
a.dropna(axis=1)
```

```python
# need correction
a.dropna(thresh=2)
```

```python
# need correction
a.dropna(axis=1,thresh=2)
```

```python
a.fillna(value='FULL VALUE')
```

```python
a.fillna(value=a.mean())
```

```python
a['C'].fillna(value=a['C'].mean())
```

```python
a
```

```python
a.fillna(value=a.max())
```

```python
a.fillna(value=a.min())
```

```python
a.iloc[4].fillna(value=a.iloc[4].min())
```
