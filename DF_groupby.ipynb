# DF.groupby

```python
import pandas as pd
```

```python
data={'Subject'      :['MATHS','MATHS','PHY','PHY','ENG','ENG'],
      'Student'      :['Ram','Renu','Kavya','Sam','Anu','Veer'],
      'External Mark':[63,78,49,39,92,70],
      'Internal Mark':[15,18,10,8,19,14]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Subject')
```

```python
by_Sub=df.groupby('Subject')
```

```python
by_Sub.sum()
```

```python
by_Sub.sum(numeric_only=True)
```

```python
df.groupby('Subject').max()
```

```python
df.groupby('Subject').max(numeric_only=True)
```

```python
by_Sub.mean(numeric_only=True)
```

```python
by_Sub.std(numeric_only=True)
```

```python
by_Sub.count()
```
