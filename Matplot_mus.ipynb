```python
import matplotlib.pyplot as plt
```

```python
# need to use this line to see the plots in the notebook :
%matplotlib inline
```

```python
# The data we want to plot :
import numpy as np
x=np.random.randint(0,50,10)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here');
plt.show()
```

```python
plt.plot(y,x)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here');
plt.show()
```

```python
plt.plot(x,y,y,x)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here');
plt.show()
```

```python
plt.plot(x,y,y,x,x**2,y**2)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here');
plt.show()
```
